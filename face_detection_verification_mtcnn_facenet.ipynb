<h4 style= 'color:#4f017f'><strong>What is MTCNN?</strong> </h4>
   <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <br><li><strong>MTCNN</strong> stands for <strong>Multi-Task Cascaded Convilutional Neural Network</strong> is DL framework which is used for Face Detection.It can detect organs of a face  like mouth , nose and ... in an image to detect facial features.MTCCN has shown more reliable detection compared with other methods in most projects. </li>
        </ul>
<h5 style= 'color:#4f017f'><strong>What are the layers of MTCNN?</strong> </h5>

<ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <br><li><strong>1- Resizing: </strong> Resizing happens multiple times to detect faces in different sizes of image.</li>
       <br><li><strong>2- P-Network & Proposal Network: </strong> This is the first step of detection and it scans proposes candidate windows thay could contain faces , therefore many false positive could be seen.These proposed images which contain erros are ready for the second Network.</li>
         <br><li><strong>3- R-Network & Refine Network: </strong>This stage is filtering all proposed images by NMS method and find great boxes which contain more reliable face frame. However , there also a number of boxes and must be just one box which is done in the last Network. </li>
         <br><li><strong>4- O-Network & Output Network: </strong>Final stage which must be determined the final box as output and take the best box amongst the boxes which have been found from the previous Netwrok.This box is exact and reliable and contains the face in the best shape.It also finds facial landmarks which is five spots of the face.These 5 spots are both eyes , nose , and both parts of mouth.</li>
        </ul>
<h5 style= 'color:#4f017f'><strong>What are Image Pyramid & NMS and Cascading  ?</strong> </h5>
<ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <br><li><strong>Image Pyramid :</strong>Actually MTCNN uses this to manage faces from different sizes.The input image is scaled to sevral resolutions and then process occures to detect faces </li>
        <br><li><strong>NMS </strong> stands for <strong>Non Maximum Suppression </strong> is used in each step of MTCNN to remove overlapping bounding boxes and get the best detection which is more reliable.</li>
         <br><li><strong>Cascading :</strong>This structure is used to make progress based on hierarchy approach.Due to each step is filtering the best boxes. </li>
        </ul>

         <li>For creating our Face Detection model a simple way I use . First , I create a function which gets a number of parameters as you wish such as image or url which for images and if i want to load an image from web , the own model , color of the box , and size of the spots.Just , image and url cant be both passed True , then  read the image by matplotlib or url by io and make a instance of our model and using detect_faces for detection and now we have the coordination of the box which is x , y , width and height . Now we just create a rectangle based on these coordinates for Face detection and for spots there is a value named keypoints in coordinate which contains these 5 spots , after we get them we draw on the image.Color class contains a number of colors based on their hash code which we can use to colorize our text or bold them.</li>
        </ul>

```python
class Color:
   GREEN = '\033[92m'
   BLUE = '\033[94m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   END = '\033[0m'
```

```python
def Face_Detection(image=None , model = MTCNN() , color ='red' , url = None , sz=10 ):
    print(f'{Color.BOLD}The process has been started for detection faces ...')
    plt.style.use('dark_background')
    if url:
        img = io.imread(url)
    if image:
        img = plt.imread(image)
    model = model
    coordinates = model.detect_faces(img)

    plt.style.use('dark_background')
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title('Face Detection')
    plt.axis('off')
    ax=plt.gca()
    for coordinate in coordinates:
        print('{}The box coordinates :{} {}\n{}The confidence :{} {}\n{}The keypoints :{} {}'.format(Color. GREEN ,Color.END ,coordinate['box'] ,Color.RED , Color.END, coordinate['confidence'],Color.BLUE , Color.END , coordinate['keypoints']))
        x,y,width , height = coordinate['box']
        rect = plt.Rectangle((x, y), width, height, fill=False, color=color)
        ax.add_patch(rect)


    plt.subplot(1, 2, 2)
    plt.imshow(img)
    plt.title(f'key points ')
    plt.axis('off')
    for coordinate in coordinates:
        left_eye = coordinate['keypoints']['left_eye']
        right_eye = coordinate['keypoints']['right_eye']
        nose = coordinate['keypoints']['nose']
        mouth_left = coordinate['keypoints']['mouth_left']
        mouth_right = coordinate['keypoints']['mouth_right']
        plt.scatter(left_eye[0] , left_eye[1], color='red' , s=sz)
        plt.scatter(right_eye[0] , right_eye[1], color='red' , s=sz)
        plt.scatter(nose[0] , nose[1], color='red' , s=sz)
        plt.scatter( mouth_left[0] , mouth_left[1], color='red' , s=sz)
        plt.scatter(mouth_right[0], mouth_right[1], color='red' , s=sz)


    plt.show()
    print('--'*30)
```

# <p style="padding:10px;background-color:#402105 ;margin:0;color:#E2EEF3;font-family:newtimeroman;font-size:100%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">MTCNN Model For Face Verification</p>

<div style="background-color:#fee9d6; padding: 20px; border-radius: 10px; box-shadow: 0 2px 4px 0 rgba(0, 0, 0, 0.1);">
    <p style="font-size:20px; font-family:calibri; color:#4f017f;"></p>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>Here , we create a function to do Detection for out image , so we create an object using MTCCN class , then using detect_faces to detect faces.Now , if our model could find any face in the image ,we get the bunding box and draw it and if not return None.Now we get the box which contains of x , y , width and height.Here we are able to use crop with a percentage of margin and detremine the min and max of both x and y based on the margin thatwe through , and get the from ymin till ymax and xmin till xmax.The final thing to do is that converting the image to the standard size which is 160 * 160 .  </li>
        </ul>

       
</div>

```python
def Face_Detection_FaceNet(image , margin=0.2):
    model = MTCNN()
    detection = model.detect_faces(image)
    if detection:
        x , y ,width , height = detection[0]['box']
        margin = margin
        x_min, y_min = max(x - int(margin * width), 0), max(y - int(margin * height), 0)
        x_max, y_max = min(x + width + int(margin * width), image.shape[1]), min(y + height + int(margin * height), image.shape[0])
        face_box  = image[y_min:y_max, x_min:x_max]
        image_with_face_detection= cv2.resize(face_box, (160, 160))
        return image_with_face_detection
    else:
        return None
```

        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
         <li>This is the shape of FaceNet model which contains a few deepl layers and then Embedding which is Face Represendation.It is a vector of features  for the face of the image and based on that we are able to calculate the loss.All we want to have is a great vector of features because in even Face Recognition which might be many people in the database all we save is a vector  of featurefor each one and based on that features we predict who he or she is .Here , we found the face in the image and we need Embedding of the face which we can use embeddings to do it for us. </li>
         <br><li><strong>Triplet loss :</strong> In a nutshell , is an algorithm which calsulates loss amongst a number of images and we aim that those images from similar group have fewer loss anddifferent images have bigger loss. In this algorithm there are 3 images , anchor , positive and negative.Anchor is the image which is the target,positive is the person the samein the  anchor and negative who is a different person. we aim that Anchor and Positive images have fewer loss and Anchor and negative have bigger loss.Triplet calculates this distance based on the Embedding vector. </li>
        </ul>

     <h4 style= 'color:#4f017f'><strong><span style='font-size:30px;'>&#128364;</span> What is Cosine Similarity?</strong> </h4>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <li><strong>Cosine similarity</strong> is a mathematical metric used to measure the similarity between two vectors in a multi-dimensional space, particularly in high-dimensional spaces, by calculating the cosine of the angle between them.Note that the direction is important in Cosine Similarity.</li>
        </ul>
        <h4 style= 'color:#4f017f'><strong><span style='font-size:30px;'>&#128364;</span> How does Cosine Similarity Work?</strong> </h4>
        <ul style="font-size:18px;line-height: 1.5em; font-family:calibri; color: #cc4e01">
        <li>Cosine Similarity determins similarity by Cosine of the angle between 2 vectors.In this process, first Dot Product will calculated and based on that the Direction of them will be found.After that the length of each one of them will be calculated.In fact the cosine similarity is the dot product divided by the product of the two vectors lengths.</li>
        </ul>

      
</div>

<center>
<img src="https://miro.medium.com/v2/resize:fit:824/1*GK56xmDIWtNQAD_jnBIt2g.png" width=1000 height=400 />

```python
def Similarity_Measurement(embedding_vector1 , embedding_vector2 , threshold = 0.7):
    Distance = cosine(embedding_vector1 ,embedding_vector2)
    scan_distnace = Distance < threshold
    return scan_distnace
```

```python
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate classification report
report = classification_report(true_labels, pred_labels, target_names=['Different', 'Same'], output_dict=True)
df_report = pd.DataFrame(report).transpose()

# Accuracy
accuracy = accuracy_score(true_labels, pred_labels)

# Prepare data
metrics_df = df_report.loc[['Different', 'Same']][['precision', 'recall', 'f1-score']]
metrics_df = metrics_df.reset_index().melt(id_vars='index', var_name='Metric', value_name='Score')
metrics_df.rename(columns={'index': 'Class'}, inplace=True)

# Set plot style
sns.set(style='whitegrid', palette='bright')

# Plot
plt.figure(figsize=(9, 6))
ax = sns.barplot(data=metrics_df, x='Class', y='Score', hue='Metric')

# Add values on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

# Title and accuracy annotation
plt.title(f'Precision, Recall, and F1-Score per Class\nAccuracy: {accuracy:.2f}', fontsize=14)
plt.ylim(0, 1.1)
plt.ylabel('Score')
plt.xlabel('')
plt.legend(title='Metric')
plt.tight_layout()
plt.show()
```
